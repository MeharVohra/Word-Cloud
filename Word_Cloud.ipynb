```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
nltk.download('punkt')
```

```python
with open("/content/drive/My Drive/JaneEyre.txt", "r") as file:
    text = file.read()
```

```python
wordcloud = WordCloud(
width=800,
height=400,
background_color="pink",
stopwords=set(STOPWORDS),
min_font_size=10,
).generate(text)
```

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
import numpy as np
from PIL import Image
```

```python
mask_path = "/content/drive/My Drive/mask image.webp"  # Replace with your mask image path
```

```python
mask_image = np.array(Image.open(mask_path))
```

```python
# Generate the word cloud with the mask
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="black",
    stopwords=set(STOPWORDS),
    mask=mask_image,  # Use the mask image for shaping the word cloud
    contour_color='white',  # Optional: outline color of the mask
    contour_width=1  # Optional: thickness of the outline
).generate(text)
```

```python
# Display the word cloud
plt.figure(figsize=(5, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
